# Prediction of variation of temperature of a heated rod after sometime

```python
from pylab import *
from numpy import *
from math import *
import random

L = 1
alpha = 0.2
delta_y = 0.01

N = int(L/delta_y)
end_time = 10
dt = (delta_y**2)*alpha
M = int(end_time/dt)

T_list = [273]
for i in range(0,N-1):
    T_list.append((random.random())+273)
T_list.append(373)
T = [T_list]
y_list = [i*delta_y for i in range(0,N+1)]

for j in range(0,M):
    T_new_list = [273]
    for i in range(1,N):
        T_new = alpha*T[j][i-1] + (1-2*alpha)*T[j][i] + alpha*T[j][i+1]
        T_new_list.append(T_new)
    T_new_list.append(373)
    T.append(T_new_list)
    
for i in range(0,10000):
    plot(y_list,T[i])



        


```
