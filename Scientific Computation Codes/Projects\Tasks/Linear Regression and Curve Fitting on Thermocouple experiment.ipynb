# Thermocouple Experiment: Linear Regression and Curve Fitting


```python
from numpy import *
import pandas as pd
from pylab import *
from sklearn.linear_model import LinearRegression

x = array([99.0, 94.5, 87.5, 80.0, 75.5, 70.5, 66.5, 60.5, 56.0, 54.0])
y = array([4.36, 4.12, 3.85, 3.41, 3.19, 2.99, 2.8, 2.53, 2.28, 2.09])

linreg = LinearRegression()
    
x = x.reshape(-1,1)
linreg.fit(x,y)

predicted_y = linreg.predict(x)

scatter(x,y)
plot(x,predicted_y,color = 'red')
xlabel('Temperature Difference (k)')
ylabel('Thermo - Emf')
title('Thermo-Emf Calibrication Graph')
legend(['Least Square Fit','Experimental Data'],loc='upper left')
figtext(0.55,0.8," y(x) = 0.048x - 0.4791",fontweight="bold",size = 9)
```
