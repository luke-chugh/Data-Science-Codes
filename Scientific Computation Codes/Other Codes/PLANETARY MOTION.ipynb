# **PLANETARY MOTION**

```
from pylab import * ## Imports a plotting program.
import math


########## Initialisation ###################################################
X = 1.0    
Y = 0.
V_x = 0.0    
V_y = 1. ## Initial velocity in units of r_0/T
A_x = -X/((X**2 + Y**2)**1.5)
A_y = -Y/((X**2 + Y**2)**1.5)
Energy = 0.5*(V_x*V_x + V_y*V_y) - 1./math.sqrt(X*X + Y*Y)
time = 0.0      
end_time =7.   ## Time upto which position and velocity are to be calculated
N = 100000           ## Number of time divisions
delta_t = end_time/N     ## Time increment

X_list = [X]
Y_list = [Y]
E_list = [Energy]
time_list = [time]         

##########    Euler Algorithm  ########################################
for i in range(0,N):
	X += delta_t*V_x
	Y += delta_t*V_y
	V_x += delta_t*A_x
	V_y += delta_t*A_y
	A_x = -X/((X**2 + Y**2)**1.5)
	A_y = -Y/((X**2 + Y**2)**1.5)
	Energy = 0.5*(V_x*V_x + V_y*V_y) - 1./math.sqrt(X*X + Y*Y)
	time += delta_t
	X_list.append(X)
	Y_list.append(Y)
	E_list.append(Energy)
	time_list.append(time)
	
#### Plotting orbit #############################################

centre_x = 0.0
centre_y = 0.0	
figure(figsize=(20,10))
xlabel('X',fontweight='bold',fontsize='14')
ylabel('Y',fontweight='bold',fontsize='14')
title('ORBIT',fontweight='bold',fontsize='20')
plot(X_list,Y_list)
plot(centre_x, centre_y, marker = 'o')
show()

#xlabel('t')
#ylabel('E')
#title('Energy conservation check')
#plot(time_list,E_list)
#show()
```
