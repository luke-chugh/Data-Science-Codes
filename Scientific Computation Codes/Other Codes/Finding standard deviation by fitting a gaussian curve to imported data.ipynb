# Fitting a Gaussian curve to data imported from time period of 1000 pendulums and hence finding standard deviation from the gaussian

```python
from matplotlib import pyplot as plt
import numpy as np
import math
file = np.loadtxt('Values.txt')
def gaussian(x):
    return 1./(math.sqrt(2.*np.pi)*0.16057)*np.exp(-np.power((x - 14.69)/0.16057, 2.)/2)
gauss = []
for i in file :
    gauss.append(1920*gaussian(i))
histogram = plt.hist(file,bins = 12,facecolor='fuchsia', edgecolor='k')
plt.plot (file,gauss)
#, density = True, facecolor='fuchsia', edgecolor='b', alpha = 0.5
plt.show()
```
