# Phase potraits of a particle under the influence of external force

```python
from pylab import *
from numpy import *
from scipy.integrate import odeint as odeint

def x(t):
    return a*exp(E[0]*t)*V[0,0] + b* exp(E[1]*t)*V[0,1]

def y(t):
    return a*exp(E[0]*t)*V[1,0] + b* exp(E[1]*t)*V[1,1] 

A=array([[-0.6,-1.2],[0.4,-0.2]])
E,V=eig(A)


a,b=randint(-10,10),randint(-10,10)
t=arange(-pi/2, pi/2, 0.01) 
plot(x(t),y(t))
plot(1.2,0.4,'o',markersize=10,color='blue')

x1, x2 = meshgrid(arange(-20, 20, 0.1), arange(-20, 20, 0.1))
xdot = A[0,0]*x1 + A[0,1]*x2
ydot = A[1,0]*x1 + A[1,1]*x2
streamplot(x1, x2, xdot, ydot,color='turquoise')
grid()

xlim(-20,20)
ylim(-20,20)
title("For $(6/5,2/5)$")
xlabel("x(t)")
ylabel("y(t)")
legend(["Trajectory of the particle","Critical Point"])
```
