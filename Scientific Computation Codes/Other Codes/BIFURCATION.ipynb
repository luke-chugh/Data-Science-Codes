# **BIFURCATION OF A HARMONIC PENDULUM**

```
from pylab import *
import math
tau = 0
t_0= 2*pi
f_0=0.5
v=0
q=0.5
a=f_0*t_0**2
b=q*t_0
omega=2./3.
kappa=omega*t_0
theta = 0.0
tau_final = 1000.0
N=10000
s=1./1000.
flist=[f_0]
dtau= tau_final/N
tau_list=[tau]
theta_list=[theta]
v_list=[v]
def f(theta,tau,v):
    return v
def g(theta,tau,v):
    return -4*pi**2*sin(theta)-b*v+a*sin(kappa*tau)
for d in range(0,1000):
    tau=0.0
    a=f_0*t_0**2
    theta =0.0  
    c=400
    for k in range(0,N):
        k1=dtau*f(theta,tau,v)
        h1=dtau*g(theta,tau,v)
        k2=dtau*f(theta+k1/2,tau+dtau/2,v+h1/2)
        h2=dtau*g(theta+k1/2,tau+dtau/2,v+h1/2)
        k3=dtau*f(theta+k2/2,tau+dtau/2,v+h2/2)
        h3=dtau*g(theta+k2/2,tau+dtau/2,v+h2/2)
        k4=dtau*f(theta+k3,tau+dtau,v+h3)
        h4=dtau*g(theta+k3,tau+dtau,v+h3)
        v=v+(h1+2*h2+2*h3+h4)/6.0
        theta=theta+(k1+2*k2+2*k3+k4)/6.0
        if(theta>math.pi):
            while(theta>math.pi):
                theta=theta-2*math.pi
        elif(theta<-math.pi):
            while(theta<-math.pi):
                theta=theta+2*math.pi
        if(tau>tau_final/2) and (abs(tau-(c/omega))<(dtau/2.)):
            tau_list.append(tau)
            theta_list.append(theta)
            v_list.append(v)
            flist.append(f_0)
            c+=1    
        tau+=dtau
    f_0+=s
plt.figure(figsize=(20,10))
plt.plot(flist,theta_list, '.', markersize = 1)
plt.title('BIFURCATION',fontweight='bold',fontsize='20')
plt.xlabel("F",fontweight='bold',fontsize='14')
plt.ylabel("THETA",fontweight='bold',fontsize='14')
plt.show()
```
