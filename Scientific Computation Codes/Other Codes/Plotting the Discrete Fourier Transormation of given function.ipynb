# Plotting the Discrete Fourier Transormation of given function

```python
from pylab import *
from math import *
from numpy import *

L = 2*pi
N = 1000 
x = linspace(0,L,N)
n = arange(0,N)

f = tan(7*x)*sin(5*x)
M = []

for i in range (0,N):
    sum = 0j
    for j in range (0,N):
        sum += exp( (-2*pi) * 1j * i * j/N ) * f[j]
    M.append(sum)
    
s = []

for i in range (0,N):
    sum = 0j
    for j in range (0,N):
        sum += exp( (2*pi) * 1j * i * j/N ) * M[j]
    s.append(sum/N)
    
plot(x,s,label="DFT Inverse")
plot(x,f,label="Orignal Fucntion")
title('$f(x) = tan(7x)sin(5x)$')
legend()
```
