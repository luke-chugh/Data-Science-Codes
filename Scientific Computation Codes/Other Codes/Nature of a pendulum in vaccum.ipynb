# Nature of a pendulum in vaccum

```python
from pylab import *
time = 0
theta = pi/6
omega = 0
time_list = [time]
theta_list = [theta]
omega_list = [omega]
E_list =[0]
N = 100000
end_time = 10
dt = end_time/N
for i in range (0,N):
    alpha = -4*pi*pi*sin(theta)
    m = 0.5*omega*omega + 4*pi*pi*(1-cos(theta))
    E_list.append(m)
    omega += alpha*dt
    theta += omega*dt
    time += dt
    omega_list.append(omega)
    theta_list.append(theta)
    time_list.append(time)
plot(time_list,omega_list,'m--')
plot(time_list,E_list,'b^')
plot(time_list,theta_list,'r-')
legend(['omega','energy','theta'],loc='upper right')
title('nature of a pendulum in vaccum')
xlabel('time')
ylabel('omega , energy , theta')
```
