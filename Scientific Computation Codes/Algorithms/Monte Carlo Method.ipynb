# **Calculation value of Pi using Monte Carlo Method**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from tqdm import tqdm

N_iter = 10000
N_trail = 1000
Pi_array = []
for i in tqdm(range(N_iter)):

    N_hit = 0
    for i in range(N_trail):
        x = np.random.random()
        y = np.random.random()


        if((x*x+y*y)<=1.000):
            N_hit += 1
            

    pi = 4.*N_hit/N_trail
    Pi_array.append(pi)

    
print("Mean Pi = ",np.mean(np.array(Pi_array)) )   
plt.figure(figsize=(20,15))
plt.xlabel('COMPUTATIONALLY GENERATED VALUE OF PI',fontweight='bold',fontsize='14')
plt.ylabel('FREQUENCY',fontweight='bold',fontsize='14')
plt.title('HISTOGRAM FOR RANDOM VALUE OF PI GENERATED',fontweight='bold',fontsize='20')
plt.hist(Pi_array,bins = 150,edgecolor='black', linewidth=1.3, facecolor='fuchsia')
plt.show()
```

# **MONTE CARLO INTEGRATION**

### Problem 1)


$$ \int_1^2  ln(x) dx $$



```python
import numpy as np ## Importing Libraries

def f(x): ## Defining integrand
    return np.log(x)

y0 = 1;a = 1;b = 2
N_trail = 10000
N_hit = 0
for i in range(N_trail):
    x = np.random.uniform(low = a,high = b)   ## Generating random numbers between the given limit
    y = np.random.uniform(low = 0,high = y0)

    if(y<f(x)): ## condition for calculating n_hits
        N_hit += 1
Y_est = (y0*(b-a)*N_hit)/N_trail ## integral value
print("Integral Value = ", Y_est)
    


```

### Problem 2)
Find The Volume inside a plane $x+y+z = 2$ in the first quadrant i.e. $$ \int_{x=0}^2\int_{y=0}^{2-x} (2-x-y)dxdy $$



I will inscribe this plane by a cube of length 2.2(seeing the upper limit) in the first quadrant then i will generate 3 random numbers simultaneously i.e. $x_0,  y_0 , z_0$ and will see whether $z_0$ lies below the plane or not


```python
import numpy as np ## Importing Libraries

def Z(x,y): ## Defining integrand which is a function of two variables
    return 2-x-y;

cube_side = 2.2
volume = cube_side**3 
N_trail = 40000

N_hit = 0
for i in range(N_trail):
    x = np.random.uniform(low = 0,high = cube_side)   ## Generating random numbers between the given limit
    y = np.random.uniform(low = 0,high = cube_side)
    z = np.random.uniform(low = 0,high = cube_side)
    if(z<Z(x,y)): ## condition for calculating n_hits
        N_hit += 1
Y_est = (volume*N_hit)/N_trail ## integral value
print("Integral Value = ", Y_est)
    


```

Clearly we see we don't get the same result every time so we take large number of data and take it's average for example  below i have plotted the histogram of the integral value that is obtained for same N_trail for N_iter times 

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm


def Z(x,y):
    return 2-x-y

cube_side = 2.2
volume = cube_side**3

N_trail = 100
N_iter = 5000

S = []

for i in tqdm(range(N_iter)):
    N_hit = 0
    for i in range(N_trail):
        x = np.random.uniform(low = 0,high = cube_side)
        y = np.random.uniform(low = 0,high = cube_side)
        z = np.random.uniform(low = 0,high = cube_side)

        if(z<Z(x,y)):
            N_hit += 1
    I_est = (volume*N_hit)/(N_trail)
    S.append(I_est)
S = np.array(S)

plt.figure(figsize = (12,10))
plt.hist(S,bins = 150,edgecolor='black', linewidth=1.2, facecolor='fuchsia')
plt.show()
print(" Integral Mean",np.mean(S))
```

---------------------
As we can see  this estimate of integral becomes precise as  $N\rightarrow \infty$ and will eventually
converge to the correct answer. This technique is an example of a ‘Direct Sampling’ Monte Carlo method!
