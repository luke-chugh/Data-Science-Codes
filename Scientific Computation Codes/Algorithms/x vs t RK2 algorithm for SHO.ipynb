# Position vs Time prediction for a simple harmonic oscillator using Runge-Kutta Algorithm

```python
from pylab import *
x = 1.0
v = 0.0
t = 0.0
E_abs = (0.5*v*v) + 0.5*(4*pi*pi)*( x*x )
dt = 0.01
n = 1000
end_time = n*dt
x_list = [x]
v_list = [v]
t_list = [t]
E_list = [E_abs]

def f(x,v,t):
    return v
def g(x,v,t):
    return (-4*pi*pi*x)

for i in range(0,n):
    k1 = dt*f(x,v,t)
    k2 = dt*g(x,v,t)
    x_new = x + dt*f(x+(k1/2),v+(k2/2),t+(dt/2))
    v = v + dt*g(x+(k1/2),v+(k2/2),t+(dt/2))
    x = x_new
    E = (0.5*v*v) + 0.5*(4*pi*pi)*( x*x )
    E_relativeerror = abs(E-E_abs)/E_abs
    t = t + dt
    x_list.append(x)
    v_list.append(v)
    t_list.append(t)
    E_list.append(E_relativeerror)
    
plot(t_list,x_list,marker = 'o' ,label = 'E_rk2' , markersize = 0.5 )
xlabel('time')
ylabel('position')
title('Position vs Time')
legend()
```
