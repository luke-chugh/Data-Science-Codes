# **DRIVEN PENDULUM USING RUNGE-KUTTA 2 ALGORITHM**

```
from pylab import *
theta  = pi/6
omega = 3
tau = 0
b =  0.5
a =  0.2
lamda = 1
theta_list = [theta]
omega_list = [omega]
tau_list = [tau]
def f(theta,omega,tau):
    return omega
def g(theta,omega,tau):
    return -4*pi*pi*theta - b*omega + a*cos(2*pi*lamda*tau)
end_time = 30
N = 300
dtau = end_time/N
for i in range(0,N):
    k1 = dtau*omega
    k2 = dtau*g(theta,omega,tau)
    theta_new = theta + dtau*f( theta+(k1/2) , omega+(k2/2) , tau+(dtau/2) ) 
    omega += dtau*g( theta+(k1/2) , omega+(k2/2) , tau+(dtau/2) )
    theta = theta_new
    tau += dtau
    theta_list.append(theta)
    omega_list.append(omega)
    tau_list.append(tau)
figure(figsize=(20,10))
plot(tau_list,theta_list,'r-')
title('THETA VS TIME IN DRIVEN HARMONIC PENDULUM RK2',fontweight='bold',fontsize='20')
xlabel('TIME',fontweight='bold',fontsize='14')
ylabel('THETA',fontweight='bold',fontsize='14')

    
```

# **DRIVEN PENDULUM USING RUNGE-KUTTA 4 ALGORITHM**

```
from pylab import *
theta  = pi/6
omega =  3
t = 0
b =  0.3
a = 0.2
lamda = 1
theta_list = [theta]
omega_list = [omega]
t_list = [t]
def f(t,theta,omega):
    return omega
def g(t,theta,omega):
    return -4*pi*pi*sin(theta) - b*omega + a*cos(2*pi*lamda*t)
end_time = 50
N = 5000
dt = end_time/N
for i in range(0,N):
    k1 = dt*f(t,theta,omega)
    k5 = dt*g(t,theta,omega)
    k2 = dt*f( t+(dt/2) , theta + (0.5*k1) , omega + (0.5*k5) )
    k6 = dt*g( t+(dt/2) , theta + (0.5*k1) , omega + (0.5*k5) )
    k3 = dt*f( t+(dt/2) , theta + (0.5*k2) , omega + (0.5*k6) )
    k7 = dt*g( t+(dt/2) , theta + (0.5*k2) , omega + (0.5*k6) )
    k4 = dt*f( t+dt , theta +k3 , omega + k7 )
    k8 = dt*g( t+dt , theta +k3 , omega + k7 )
    theta += (1/6)*( k1 + 2*k2 + 2*k3 + k4 )
    omega += (1/6)*( k5 + 2*k6 + 2*k7 + k8 )
    t += dt
    theta_list.append(theta)
    omega_list.append(omega)
    t_list.append(t)
figure(figsize=(20,10))
plot(t_list,theta_list,'r-')
title('THETA VS TIME IN DRIVEN HARMONIC PENDULUM RK4',fontweight='bold',fontsize='20')
xlabel('THETA',fontweight='bold',fontsize='14')
ylabel('OMEGA',fontweight='bold',fontsize='14')
```
