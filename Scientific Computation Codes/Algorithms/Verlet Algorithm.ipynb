# **Simple Pendulum by Verlet Algorithm**

```python
from ipywidgets import*  ##importing some libraries
import matplotlib.pyplot as plt 
from IPython.display import display
import numpy as np

def f_main (theta,plot):    
    th = theta*np.pi/180
    w = 0.0 ##initial angular velocity##
    a = -np.sin(th) ## angular acceleration##
    t = 0.0  ##initial time##
    energy = w*w+ 2*(1-np.cos(th))  
    et = 21 ##end_time
    N = 10000
    dt = (et-t)/N  ###increment in time
    W = [w]
    Th = [th]    ##create theeta lists
    time_list = [t]  ##create time list
    e_list = [energy]  ###create energy  list
    if(plot==True):
        
        for i in range(1,N):
            
            ##updation using verlet algorithm
            th += w*dt/2
            a = -np.sin(th)
            w+= a*dt
            th += w*dt/2
            
            t+=dt
            energy = w*w+ 2*(1-np.cos(th))

            Th.append(th)  ##stroing them in a list
            time_list.append(t)
            e_list.append(energy)
            W.append(w)

        #plotting list in three different graphs 

        fig,((ax1),(ax2),(ax3)) = plt.subplots(nrows = 3,ncols= 1,figsize = (15,25))
        ax1.set_title(r'$\theta \ \ vs \ \ \tau$',fontweight='bold',fontsize='20')
        ax2.set_title(r'$\omega \ \ vs \ \ \tau$',fontweight='bold',fontsize='20')
        ax3.set_title(r'$E \ \ vs \ \ \tau$',fontweight='bold',fontsize='20')

        ax1.plot(time_list,Th,color = 'b',label='Theta')
        ax2.plot(time_list,W,'k',label = 'energy')
        ax3.set_ylim(0,2.2)
        ax3.plot(time_list,e_list,color = 'r',label = 'energy')
        plt.legend()

        plt.show()


#####creating a widget that will communiccate with our f_main function and provide them different values
w = interactive(f_main,theta = IntSlider(min = -90,max = 90,value = 1,description = "Theeta(0) in Degrees",layout = {'width':'700px'}),plot = False)
###displaying it and finally using display :')
display(w)
```
